```python
# Шаг 1: Загрузка данных и предварительный анализ
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Загрузка исходного набора данных
dataset_path = 'website_traffic_data(Dataset2).csv'
dataset = pd.read_csv(dataset_path)

# Удаление дубликатов и пропущенных значений
dataset = dataset.drop_duplicates().dropna()

# Для второй модели используем только один входной параметр 'Conversion Rate' и целевую переменную 'Bounce Rate'
dataset = dataset[['Conversion Rate', 'Bounce Rate']]

# Нормализуем столбец 'Conversion Rate'
scaler = StandardScaler()
dataset['Conversion Rate'] = scaler.fit_transform(dataset[['Conversion Rate']])
```

```python
# Шаг 2: Визуализация данных
```

```python
# Гистограммы
plt.figure(figsize=(12, 8))
numeric_columns = ['Conversion Rate', 'Bounce Rate']
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(1, 2, i)  # 1 строка, 2 столбца
    sns.histplot(dataset[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Распределение {col}')
plt.tight_layout()
plt.show()

# Корреляционная матрица
plt.figure(figsize=(8, 6))
sns.heatmap(dataset[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляционная матрица')
plt.show()
```

```python
# Шаг 3: Сохранение подготовленных данных
```

```python
dataset.to_csv('dataset_2.csv', index=False)

# Подтверждение сохранения
print("Данные успешно сохранены.")
```
